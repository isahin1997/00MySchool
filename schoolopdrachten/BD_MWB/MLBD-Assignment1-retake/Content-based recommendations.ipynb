## Preprocess data

Hiervoor zullen we eerst de data opschonen. Dit is belangrijk omdat strings vaak informatie bevat die niet relevant zijn voor het systeem. Denk hier aan informatie zoals:

- Punctuaties -> ? ! , .
- Stopwoorden
- Geen alfabetische karakters

Deze informatie zegt weinig over de context van een stuk test. Daarom zal dit er ook uit gefilterd worden.

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
import re
nltk.download('stopwords')

data = df.loc[:,['Plot', 'Title', 'Writer']]
result = df.loc[:,['Plot', 'Title', 'Writer']]
```

Verder is er een ander probleem binnen de data. Het komt vaak voor dat er veschillende vormen bestaan van een woord. Voorbeeld:

- Werken
- Werk
- Werkte

Al deze vormen hebben dezelfde betekenis, maar wordt op dit moment door het systeem gezien als verschillende woorden. Het is belangrijk dat deze woorden als dezelfde woord worden gezien door het systeem. Dit kan gedaan worden door middel van Stemming. Stemming zorgt ervoor dat de woorden worden omgezet naar een standaardvorm. voorbeeld:

- Werken -> Werk
- Werk -> Werk
- Werkte -> Werk

```python
# Eerst zullen we functies definieren voor data cleaning. In deze stap willen we irrelevante informatie verwijderen uit de data.
# hiermee bedoelen we verwijderen van punctions, karakters buiten de alfabet, stopwoorden.
# daarnaast gebruiken we stemming voor....

def cleanPunc(sentence): #function to clean the word of any punctuation or special characters
    cleaned = re.sub(r'[?|!|\'|"|#]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    cleaned = cleaned.strip()
    cleaned = cleaned.replace("\n"," ")
    return cleaned


def keepAlpha(sentence):
    alpha_sent = ""
    for word in sentence.split():
        alpha_word = re.sub('[^a-z A-Z]+', ' ', word)
        alpha_sent += alpha_word
        alpha_sent += " "
    alpha_sent = alpha_sent.strip()
    return alpha_sent

stop_words = set(stopwords.words('english'))
re_stop_words = re.compile(r"\b(" + "|".join(stop_words) + ")\\W", re.I)


def removeStopWords(sentence):
    global re_stop_words
    return re_stop_words.sub(" ", sentence)

stemmer = SnowballStemmer("english")
def stemming(sentence):
    stemSentence = ""
    for word in sentence.split():
        stem = stemmer.stem(word)
        stemSentence += stem
        stemSentence += " "
    stemSentence = stemSentence.strip()
    return stemSentence

data['Plot'] = data['Plot'].str.lower() # upper case -> lower case.
data['Plot'] = data['Plot'].apply(cleanPunc) # remove punctuations ? . ! , ; etc..
data['Plot'] = data['Plot'].apply(keepAlpha) # remove numerical values. use only alphabetical characters.
data['Plot'] = data['Plot'].apply(removeStopWords) # remove stop words.
data['Plot'] = data['Plot'].apply(stemming) # stemming words

data['Title'] = data['Title'].str.lower() # upper case -> lower case.
data['Title'] = data['Title'].apply(cleanPunc) # remove punctuations ? . ! , ; etc..
data['Title'] = data['Title'].apply(keepAlpha) # remove numerical values. use only alphabetical characters.
data['Title'] = data['Title'].apply(removeStopWords) # remove stop words.
data['Title'] = data['Title'].apply(stemming) # stemming words

data['Writer'] = data['Writer'].str.lower() # upper case -> lower case.
data['Writer'] = data['Writer'].apply(cleanPunc) # remove punctuations ? . ! , ; etc..
data['Writer'] = data['Writer'].apply(keepAlpha) # remove numerical values. use only alphabetical characters.
#df['Writer'] = df['Writer'].apply(removeStopWords) # remove stop words.


print('-----------Done-------------')
```

Voor namen zal er geen stemming worden toegepast, omdat dit juist het omgekeerde effect zal leveren. Een naam is aardig uniek en heeft geen verschillende vormen met dezelfde betekenis. Daarom zal het hier niet toegepast worden.

Nu we de data hebben opgeschoond kunnen we de dataset omzetten naar numerieke waardes. De reden waarom we dit doen, is omdat een computer meer begrijkt van cijfers, dan van woorden. Een computer kan goed met numerieke waardes omgaan. Hiervoor bestaan ook verschillende technieken.

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

# using different vectorizers
vectorizer_title = TfidfVectorizer(strip_accents='unicode', analyzer='word', ngram_range=(1,2), norm='l2')
vectorizer_plot = TfidfVectorizer(strip_accents='unicode', analyzer='word', ngram_range=(1,3), norm='l2')
vectorizer_writer = CountVectorizer()

vectorizer_title.fit(data['Title'])

vectorizer_plot.fit(data['Plot'])

vectorizer_writer.fit(data['Writer'])

x_train_plot = vectorizer_plot.transform(data['Plot'])
bag_of_words_writer = vectorizer_writer.transform(data['Writer'])
bag_of_words_title = vectorizer_title.transform(data['Title'])
```

Bag of Words is voldoende om gebruikt te worden voor de schrijver. TF-IDF is namelijk een meer geavenceerde vectorizer die gebruik wordt voor stukken tekst waarin zeldzame woorden gevonden moeten worden die veel informatie bevat over de context. 

Hieronder definieren we de aantal aanbevelingen die we willen doen met de applicatie

```python
# set amount of recommendations that you want
amount_recommendations = 5
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

#selected_movie = 7842
# select movie
#movie = data.loc[selected_movie]

def find_recommendations(m, index_movie):
    
    selected_movie = index_movie
    
    # choose a movie
    movie_plot = vectorizer_plot.transform([m['Plot']])
    movie_title = vectorizer_title.transform([m['Title']])
    movie_writer = vectorizer_writer.transform([m['Writer']])

    similarity_plot = []
    similarity_title = []
    similarity_writer = []

    # movie plot
    print('start for plots')
    for i in range(0, len(data)):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_plot, x_train_plot[i-1000 : i].toarray())
            similarity_plot = np.append(similarity_plot, c_sim)
        elif i == (len(data) - 1):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_plot, x_train_plot[i-leftover : i+1].toarray())
            similarity_plot = np.append(similarity_plot, c_sim)
    print('plot similarity is done')
    # movie writer
    print('start for writers')
    for i in range(0, len(data)):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_writer, bag_of_words_writer[i-1000 : i].toarray())
            similarity_writer = np.append(similarity_writer, c_sim)
        elif i == (len(data) - 1):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_writer, bag_of_words_writer[i-leftover : i+1].toarray())
            similarity_writer = np.append(similarity_writer, c_sim)
    print('writer similarity is done')
    # movie title
    print('start for titles')
    for i in range(0, len(data)):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_title, bag_of_words_title[i-1000 : i].toarray())
            similarity_title = np.append(similarity_title, c_sim)
        elif i == (len(data) - 1):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_title, bag_of_words_title[i-leftover : i+1].toarray())
            similarity_title = np.append(similarity_title, c_sim)
    print('title similarity is done')
    print('--------------Done--------------')
    global result
    tmp_result = result
    
    tmp_result['c_sim_plot'] = similarity_plot
    tmp_result['c_sim_title'] = similarity_title
    tmp_result['c_sim_writer'] = similarity_writer
    tmp_result['c_sim_total'] = np.add( np.add(tmp_result['c_sim_plot'] * 10, tmp_result['c_sim_title']), tmp_result['c_sim_writer'] * 0.6)
    sorted_list = tmp_result.sort_values('c_sim_total', ascending=False)
    global amount_recommendations
    sorted_list.drop(selected_movie, inplace=True)
    return sorted_list.head(amount_recommendations)
```

## Toelichting merge strategie

Om de aanbevelingen te doen, is er een merge strategie uitgevoerd voor de overeenkomsten die zijn gevonden. Allereerst hebben we per kolom de overeenkomsten gevonden. Na wat knoeien met de parameters ziet de merge strategie er als volgt uit.

#### Plot

De samenvatting tellen we zwaarder mee. Dit komt omdat er tussen samenvattingen weinig overeenkomsten bestaan. De gelijkheden zullen voornamelijk in zeldzame woorden zitten zoals een naam van een superheld. Dit komt zelden voor, maar als het voorkomt dan is de kans groter dat de films iets met elkaar te maken hebben

#### Title

De titel behoudt zijn eigen waarde. Het wordt niet vermenigvuldigd met een bepaalde factor. Overeenkomsten in titels vinden is zeer sterk. Voornamelijk voor films die een blockbuster zijn. Een voorbeeld is Harry Potter.

De reden waarom het geen factor heeft gekregen is omdat we het niet te zwaar willen meetellen. Een voorbeeld is de film franchise Lord of The Rings. Toevallig bestaat er een andere film die The Ring heet. Door een positieve factor mee te geven kwam het voor dat The Ring eerder werd aangeraden dan The Hobbit (Prequel van The Lord of The Rings).

### Writer

De schrijver rijgt een negatieve factor. Het komt vaak voor dat efilms dezelfde schrijvers hebben, maar de films qua onderwerp of genre niks met elkaar te maken hebben. Door deze factor verkleinen we dus deze kans.

```python
def retrieve_similar_movies(m, index_movie):
    
    selected_movie = index_movie
    
    print('---searching for movie similarities---')
    
    # choose a movie
    movie_plot = vectorizer_plot2.transform([m['Plot']])
    movie_title = vectorizer_title2.transform([m['Title']])
    movie_writer = vectorizer_writer2.transform([m['Writer']])
    movie_director = vectorizer_Director2.transform([m['Director']])
    movie_genre = vectorizer_Genre2.transform([m['Genre']])
    
    similarity_plot = []
    similarity_title = []
    similarity_writer = []
    similarity_director = []
    similarity_genre = []
    # movie plot
    print('start for plots')
    for i in range(0, len(data2) +1):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_plot, x_train_plot[i-1000 : i].toarray())
            similarity_plot = np.append(similarity_plot, c_sim)
        elif i == (len(data2)):
            leftover = i % 1000
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_plot, x_train_plot[i-leftover : i].toarray())
            similarity_plot = np.append(similarity_plot, c_sim)
    print('plot similarity is done')
    # movie writer
    print('start for writers')
    for i in range(0, len(data2) +1):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_writer, bag_of_words_writer[i-1000 : i].toarray())
            similarity_writer = np.append(similarity_writer, c_sim)
        elif i == (len(data2)):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_writer, bag_of_words_writer[i-leftover : i].toarray())
            similarity_writer = np.append(similarity_writer, c_sim)
    print('writer similarity is done')
    # movie title
    print('start for titles')
    for i in range(0, len(data2) + 1):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_title, bag_of_words_title[i-1000 : i].toarray())
            similarity_title = np.append(similarity_title, c_sim)
        elif i == (len(data2)):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_title, bag_of_words_title[i-leftover : i].toarray())
            similarity_title = np.append(similarity_title, c_sim)
    print('title similarity is done')
    # movie director
    print('start for directors')
    for i in range(0, len(data2) + 1):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_director, bag_of_words_director[i-1000 : i].toarray())
            similarity_director = np.append(similarity_director, c_sim)
        elif i == (len(data2)):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_director, bag_of_words_director[i-leftover : i].toarray())
            similarity_director = np.append(similarity_director, c_sim)
    print('director similarity is done')
    # movie genre
    print('start for genre')
    for i in range(0, len(data2)+ 1):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_genre, bag_of_words_genre[i-1000 : i].toarray())
            similarity_genre = np.append(similarity_genre, c_sim)
        elif i == (len(data2)):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_genre, bag_of_words_genre[i-leftover : i].toarray())
            similarity_genre = np.append(similarity_genre, c_sim)
    print('genre similarity is done')
    print('--------------Done--------------')
    global result2
    tmp_result = result2
    tmp_result['c_sim_plot'] = similarity_plot
    tmp_result['c_sim_title'] = similarity_title
    tmp_result['c_sim_writer'] = similarity_writer
    tmp_result['c_sim_director'] = similarity_writer
    tmp_result['c_sim_genre'] = similarity_genre
    tmp_result['c_sim_total'] = np.add(np.add(np.add(np.add(tmp_result['c_sim_genre'] * 0.1, tmp_result['c_sim_director'] * 0.5), tmp_result['c_sim_plot'] * 10), tmp_result['c_sim_title'] * 2), tmp_result['c_sim_writer'] * 0.25)
    
    global selection_movies
    tmp_result = tmp_result.drop(selection_movies)
    return tmp_result

def find_recommendations2(movies):
    recommended_movies2 = pd.DataFrame()
    for i in range(0, len(movies)):
        movie = movies.iloc[i]
        print('----- similarities for movie ', i, '------')
        retrieved_result = retrieve_similar_movies(movie, movie.index)
        recommended_movies2 = recommended_movies2.append(retrieved_result)
        #pd.concat([recommended_movies2, retrieved_result])
        
    print('recommended movies')
    recommended_movies2.drop_duplicates(subset=['imdbId'], inplace=True)
    sorted_list = recommended_movies2.sort_values('c_sim_total', ascending=False)
    global amount_recommendations
    return sorted_list.head(amount_recommendations)
```
