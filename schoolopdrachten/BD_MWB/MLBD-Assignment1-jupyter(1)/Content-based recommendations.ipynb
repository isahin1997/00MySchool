```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
import re
nltk.download('stopwords')

data = df.loc[:,['Plot', 'Title', 'Writer']]
result = df.loc[:,['Plot', 'Title', 'Writer']]
```

```python
# Eerst zullen we functies definieren voor data cleaning. In deze stap willen we irrelevante informatie verwijderen uit de data.
# hiermee bedoelen we verwijderen van punctions, karakters buiten de alfabet, stopwoorden.
# daarnaast gebruiken we stemming voor....

def cleanPunc(sentence): #function to clean the word of any punctuation or special characters
    cleaned = re.sub(r'[?|!|\'|"|#]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    cleaned = cleaned.strip()
    cleaned = cleaned.replace("\n"," ")
    return cleaned


def keepAlpha(sentence):
    alpha_sent = ""
    for word in sentence.split():
        alpha_word = re.sub('[^a-z A-Z]+', ' ', word)
        alpha_sent += alpha_word
        alpha_sent += " "
    alpha_sent = alpha_sent.strip()
    return alpha_sent

stop_words = set(stopwords.words('english'))
re_stop_words = re.compile(r"\b(" + "|".join(stop_words) + ")\\W", re.I)


def removeStopWords(sentence):
    global re_stop_words
    return re_stop_words.sub(" ", sentence)

stemmer = SnowballStemmer("english")
def stemming(sentence):
    stemSentence = ""
    for word in sentence.split():
        stem = stemmer.stem(word)
        stemSentence += stem
        stemSentence += " "
    stemSentence = stemSentence.strip()
    return stemSentence

data['Plot'] = data['Plot'].str.lower() # upper case -> lower case.
data['Plot'] = data['Plot'].apply(cleanPunc) # remove punctuations ? . ! , ; etc..
data['Plot'] = data['Plot'].apply(keepAlpha) # remove numerical values. use only alphabetical characters.
data['Plot'] = data['Plot'].apply(removeStopWords) # remove stop words.
data['Plot'] = data['Plot'].apply(stemming) # stemming words

data['Title'] = data['Title'].str.lower() # upper case -> lower case.
data['Title'] = data['Title'].apply(cleanPunc) # remove punctuations ? . ! , ; etc..
data['Title'] = data['Title'].apply(keepAlpha) # remove numerical values. use only alphabetical characters.
data['Title'] = data['Title'].apply(removeStopWords) # remove stop words.
data['Title'] = data['Title'].apply(stemming) # stemming words

data['Writer'] = data['Writer'].str.lower() # upper case -> lower case.
data['Writer'] = data['Writer'].apply(cleanPunc) # remove punctuations ? . ! , ; etc..
data['Writer'] = data['Writer'].apply(keepAlpha) # remove numerical values. use only alphabetical characters.
#df['Writer'] = df['Writer'].apply(removeStopWords) # remove stop words.


print('-----------Done-------------')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

# using different vectorizers
vectorizer_title = TfidfVectorizer(strip_accents='unicode', analyzer='word', ngram_range=(1,2), norm='l2')
vectorizer_plot = TfidfVectorizer(strip_accents='unicode', analyzer='word', ngram_range=(1,3), norm='l2')
vectorizer_writer = CountVectorizer()

vectorizer_title.fit(data['Title'])

vectorizer_plot.fit(data['Plot'])

vectorizer_writer.fit(data['Writer'])

x_train_plot = vectorizer_plot.transform(data['Plot'])
bag_of_words_writer = vectorizer_writer.transform(data['Writer'])
bag_of_words_title = vectorizer_title.transform(data['Title'])
```

```python
# set amount of recommendations that you want
amount_recommendations = 5
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

#selected_movie = 7842
# select movie
#movie = data.loc[selected_movie]

def find_recommendations(m, index_movie):
    
    selected_movie = index_movie
    
    # choose a movie
    movie_plot = vectorizer_plot.transform([m['Plot']])
    movie_title = vectorizer_title.transform([m['Title']])
    movie_writer = vectorizer_writer.transform([m['Writer']])

    similarity_plot = []
    similarity_title = []
    similarity_writer = []

    # movie plot
    print('start for plots')
    for i in range(0, len(data)):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_plot, x_train_plot[i-1000 : i].toarray())
            similarity_plot = np.append(similarity_plot, c_sim)
        elif i == (len(data) - 1):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_plot, x_train_plot[i-leftover : i+1].toarray())
            similarity_plot = np.append(similarity_plot, c_sim)
    print('plot similarity is done')
    # movie writer
    print('start for writers')
    for i in range(0, len(data)):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_writer, bag_of_words_writer[i-1000 : i].toarray())
            similarity_writer = np.append(similarity_writer, c_sim)
        elif i == (len(data) - 1):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_writer, bag_of_words_writer[i-leftover : i+1].toarray())
            similarity_writer = np.append(similarity_writer, c_sim)
    print('writer similarity is done')
    # movie title
    print('start for titles')
    for i in range(0, len(data)):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_title, bag_of_words_title[i-1000 : i].toarray())
            similarity_title = np.append(similarity_title, c_sim)
        elif i == (len(data) - 1):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_title, bag_of_words_title[i-leftover : i+1].toarray())
            similarity_title = np.append(similarity_title, c_sim)
    print('title similarity is done')
    print('--------------Done--------------')
    global result
    tmp_result = result
    
    tmp_result['c_sim_plot'] = similarity_plot
    tmp_result['c_sim_title'] = similarity_title
    tmp_result['c_sim_writer'] = similarity_writer
    tmp_result['c_sim_total'] = np.add( np.add(tmp_result['c_sim_plot'] * 10, tmp_result['c_sim_title']), tmp_result['c_sim_writer'] * 0.6)
    sorted_list = tmp_result.sort_values('c_sim_total', ascending=False)
    global amount_recommendations
    sorted_list.drop(selected_movie, inplace=True)
    return sorted_list.head(amount_recommendations)
```

```python
def retrieve_similar_movies(m, index_movie):
    
    selected_movie = index_movie
    
    print('---searching for movie similarities---')
    
    # choose a movie
    movie_plot = vectorizer_plot2.transform([m['Plot']])
    movie_title = vectorizer_title2.transform([m['Title']])
    movie_writer = vectorizer_writer2.transform([m['Writer']])
    movie_director = vectorizer_Director2.transform([m['Director']])
    movie_genre = vectorizer_Genre2.transform([m['Genre']])
    
    similarity_plot = []
    similarity_title = []
    similarity_writer = []
    similarity_director = []
    similarity_genre = []
    
    # movie plot
    print('start for plots')
    for i in range(0, len(data)):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_plot, x_train_plot[i-1000 : i].toarray())
            similarity_plot = np.append(similarity_plot, c_sim)
        elif i == (len(data) - 1):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_plot, x_train_plot[i-leftover : i+1].toarray())
            similarity_plot = np.append(similarity_plot, c_sim)
    print('plot similarity is done')
    # movie writer
    print('start for writers')
    for i in range(0, len(data)):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_writer, bag_of_words_writer[i-1000 : i].toarray())
            similarity_writer = np.append(similarity_writer, c_sim)
        elif i == (len(data) - 1):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_writer, bag_of_words_writer[i-leftover : i+1].toarray())
            similarity_writer = np.append(similarity_writer, c_sim)
    print('writer similarity is done')
    # movie title
    print('start for titles')
    for i in range(0, len(data)):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_title, bag_of_words_title[i-1000 : i].toarray())
            similarity_title = np.append(similarity_title, c_sim)
        elif i == (len(data) - 1):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_title, bag_of_words_title[i-leftover : i+1].toarray())
            similarity_title = np.append(similarity_title, c_sim)
    print('title similarity is done')
    # movie director
    print('start for directors')
    for i in range(0, len(data)):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_director, bag_of_words_director[i-1000 : i].toarray())
            similarity_director = np.append(similarity_director, c_sim)
        elif i == (len(data) - 1):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_director, bag_of_words_director[i-leftover : i+1].toarray())
            similarity_director = np.append(similarity_director, c_sim)
    print('director similarity is done')
    # movie genre
    print('start for genre')
    for i in range(0, len(data)):
        if i % 1000 == 0 and i != 0:
            #print(i)
            c_sim = cosine_similarity(movie_genre, bag_of_words_genre[i-1000 : i].toarray())
            similarity_genre = np.append(similarity_genre, c_sim)
        elif i == (len(data) - 1):
            leftover = i % 1000
            #print('we are now at: ', i)
            #print('leftover: ', leftover)
            c_sim = cosine_similarity(movie_genre, bag_of_words_genre[i-leftover : i+1].toarray())
            similarity_genre = np.append(similarity_genre, c_sim)
    print('genre similarity is done')
    print('--------------Done--------------')
    global result2
    tmp_result = result2
    
    tmp_result['c_sim_plot'] = similarity_plot
    tmp_result['c_sim_title'] = similarity_title
    tmp_result['c_sim_writer'] = similarity_writer
    tmp_result['c_sim_director'] = similarity_writer
    tmp_result['c_sim_genre'] = similarity_genre
    tmp_result['c_sim_total'] = np.add(np.add(np.add(np.add(tmp_result['c_sim_genre'], tmp_result['c_sim_director']), tmp_result['c_sim_plot'] * 10), tmp_result['c_sim_title']), tmp_result['c_sim_writer'] * 0.6)
    
    global selection_movies
    tmp_result = tmp_result.drop(selection_movies)
    return tmp_result

def find_recommendations2(movies):
    recommended_movies = pd.DataFrame()
    for i in range(0, len(movies)):
        movie = movies.iloc[i]
        retrieved_result = retrieve_similar_movies(movie, movie.index)
        pd.concat([recommended_movies, retrieved_result])
        
    recommended_movies.drop_duplicates(subset=['index'], inplace=True)
    sorted_list = recommended_movies.sort_values('c_sim_total', ascending=False)
    global amount_recommendations
    return sorted_list.head(amount_recommendations)
```
